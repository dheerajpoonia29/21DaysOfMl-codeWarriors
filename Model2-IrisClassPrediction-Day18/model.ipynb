# SPLITING DATA INTO FEATURE (INDEPENDENT VAR) AND LABEL (DEPENDENT VAR)

*   x feature - independent variable
*   y label - dependent variable


*   List item
*   List item





```
x = data.iloc[:,0:4]
print(x.head())
```

```
y = data.iloc[:,-1]
print(y.head())
```

```
# above x, y in data frame
# we need to convert it into numpy array
print(type(x))

x = x.values
y = y.values

print(type(x))
```

# TRANSFORM CATEGORICAL VALUE INTO NUMERIC

```
from sklearn.preprocessing import LabelBinarizer
# label_binarize perform both one hot encoding as well as label encoding one after another

# label encoding : 0 for male, 1 for female
# one hote encoding : on nth row if male then set-1 else set-0 oor if female place 1 not femal place 0

lb = LabelBinarizer()
y = lb.fit_transform(y)
```

```
print(y[0:5, :])
```

# SPLITING DATA SET INTO TEST AND TRAIN


```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```
print('training dataset \n')
print('> ', x_train.shape, y_train.shape)

print('\ntesting dataset \n')
print('> ', x_test.shape, y_test.shape)
```

```
print(x_train.shape, y_train.shape)
print('training feature: \n', x_train[:5, :])
print('training label: \n', y_train[:5, :])
```

```
import numpy as np

# this need to done for checking accuracy 
print(y_test[0:5, :])
y_test = np.argmax(y_test, axis=1)
print(y_test)

# similarly we also do with y_pred after getting data
```

# TRAINING MODEL


```
import numpy as np
from sklearn.metrics import accuracy_score
```

```
# RANDOME FOREST - CODE WARRIOR

from sklearn.ensemble import RandomForestClassifier

rf_classifier = RandomForestClassifier(n_estimators=100)
# n_estimators -> no of tree we want to use

model = rf_classifier.fit(x_train, y_train)
print('model created successfully')
```

```
# DECISION TREE

from sklearn.tree import DecisionTreeClassifier

dt_classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)

model = dt_classifier.fit(x_train, y_train)
print('model created successfully')
```

```
# NAIVE BAYES

from sklearn.naive_bayes import GaussianNB

nb_classifer = GaussianNB()

model = nb_classifer.fit(x_train, y_train)
print('model created successfully')
```

```
# SUPPORT VECTOR MACHINE

from sklearn.svm import SVC

svm_classifier = SVC(kernal='rbf')

model = svm_classifier.fit(x_train, y_train)

print('model created successfully')
```

```
# KNN

from sklearn.neighbors import KNeighborsClassifier

knn_classifier = KNeighborsClassifier(n_neighbors=5, metric='minkowski', p=2)

model = knn_classifier.fit(x_train, y_train)

print('model created successfully')
```

```
# LOGISTIC REGRESSION

from sklearn.linear_model import LogisticRegression

lr_classifier = LogisticRegression()

model = lr_classifier.fit(x_train, y_train)

print('model created successfully')
```

# TESTING MODEL

```
y_pred = model.predict(x_test)
print(y_pred[0:5, :])
y_pred = np.argmax(y_pred, axis=1)
print(y_pred)

from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_pred, y_test)
print(cm)
# rest then diagonal are zero then we say 0% error or 100% accuracy

from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_pred, y_test)
print('accuracy: ', accuracy*100,'%')
```
