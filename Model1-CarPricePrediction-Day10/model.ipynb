# APPLYING DIFFERENT REGRESSION LEARNING ALGORITHM

````
import numpy as np
import time
````

````
# LINEAR REGRESSION / MULTIPLE LINEAR REGRESSION
start = time.clock()

from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression()

model = linear_regressor.fit(x_train, y_train)

print('model created successfully')

# y_pred = model.predict(x_test)

accuracy = model.score(x_test,y_test)
print('accuracy: ',accuracy*100,'%')

print('time taken: ', int(time.clock() - start), 'sec')
````

````
# POLYNOMIAL REGRESSION

from sklearn.preprocessing import PolynomialFeatures
start = time.clock()

polynomial_regressor = PolynomialFeatures(degree=3)

x_poly = polynomial_regressor.fit_transform(x_train)

linear_regressor = LinearRegression()

model = linear_regressor.fit(x_poly, y_train)

print('model created successfully')

accuracy = model.score(x_test,y_test)
print('accuracy: ',accuracy*100,'%')

print('time taken: ', int(time.clock() - start), 'sec')
````

````
# SUPPORT VECTOR REGRESSION 

from sklearn.svm import SVR
start = time.clock()

sv_regressor = SVR(kernel='rbf')

model = sv_regressor.fit(x_train, y_train)

print('model created successfully')

accuracy = model.score(x_test,y_test)
print('accuracy: ',accuracy*100,'%')

print('time taken: ', int(time.clock() - start), 'sec')
````

````
# DECISION TREE 

from sklearn.tree import DecisionTreeRegressor
start = time.clock()

dt_regressor = DecisionTreeRegressor(random_state=0)

model = dt_regressor.fit(x_train, y_train)

print('model created successfully')

accuracy = model.score(x_test,y_test)
print('accuracy: ',accuracy*100,'%')

print('time taken: ', int(time.clock() - start), 'sec')
````

````
# RANDOM FOREST 

from sklearn.ensemble import RandomForestRegressor
start = time.clock()

rd_regressor = RandomForestRegressor(n_estimators=10, random_state=0)

print('shape of training dataset: ',x_train.shape, y_train.shape)

model = rd_regressor.fit(x_train, y_train) 

print('model created successfully')

y_test = y_test.reshape(-1, 1)

y_pred = model.predict(x_test)

y_pred = y_pred.reshape(-1, 1)

print('shape of testing dataset: ',x_test.shape, y_test.shape)
print(y_pred.shape, y_test.shape)


accuracy = model.score(y_pred, y_test)
# print('accuracy: ',accuracy*100,'%')

print('time taken: ', int(time.clock() - start), 'sec')
````

````
# RANDOM FOREST - CODE WARRIOR

from sklearn.ensemble import RandomForestRegressor
start = time.clock()

regressor = RandomForestRegressor(n_estimators=300,random_state=0)

model = regressor.fit(x_train,y_train)

print('model created successfully')

accuracy = model.score(x_test,y_test)
print('accuracy: ',accuracy*100,'%')

print('time taken: ', int(time.clock() - start), 'sec')
````
